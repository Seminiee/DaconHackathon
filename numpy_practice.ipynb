# 브로드캐스팅(Broadcasting)의 이해

## 브로드캐스팅
- shape이 같은 두 ndarray에 대한 연산은 각 원소별로 진행
- 연산되는 두 ndarray가 다른 shape을 갖는 경우 브로드캐스팅(shape을 맞춤) 후 진행
- 뒷 차원에서 부터 비교하여 shape이 같어나, 차원 중 값이 1인 것이 존재하면 가능

![img](https://www.tutorialspoint.com/numpy/images/array.jpg)

- shape이 같은 경우

```python
x = np.arange(15).reshape(3, 5)
y = np.random.randint(0, 15, 15).reshape(3, 5)
x + y
```

- scalar(상수)와의 연산

```python
x % 2 == 0
```

```python
x * 2
```

- shape이 다른 경우 연산
    - 뒷 차원의 size가 다른 경우 -> 연산 불가

```python
a = np.arange(12).reshape(4,3)
b = np.arange(100,103)

a + b
```

```python
c = np.arange(1000, 1004)
a + c # Error
```

```python
d = b.reshape(1,3)

print('b')
print(b)
print(b.shape)
print('d')
print(d)
print(d.shape)
```

```python
a + d
```

## 다중조건 사용하기
- 파이썬 논리 연산자인 `and`, `or`, `not` 키워드 사용할 수 없음
- `&` - AND
- `|` - OR
- NOT의 경우 조건을 반대로 명시

```python
# x가 짝수면서 30보다 작은 경우
x[(x % 2 == 0) & (x < 30)]
```

```python
# x가 50보다 크거나 30보다 작은 경우
x[(x > 50) | (x < 30)]
```

# linalg 서브모듈 함수 활용하여 선형대수 연산하기

## `np.linalg.inv`
- 역행렬을 구할 때 사용
- 모든 차원의 값이 같아야 함(정방행렬만 가능)

```python
x = np.random.randint(-50, 50, size = (3,3))
x
```

```python
np.linalg.inv(x)
```

```python
x @ np.linalg.inv(x) # 행렬곱
```

```python
np.matmul(x, np.linalg.inv(x)) # 행렬곱 
```

## `np.linalg.solve`
- Ax = B 형태의 선형대수식 솔루션을 제공

```python
A = np.array([[1, 1], 
                                    [2, 4]])
B = np.array([25, 64])

np.linalg.solve(A, B)
```

```python
np.allclose(A @ np.linalg.solve(A, B), B)
```

2. numpy를 이용하여 pi(원주율)값을 계산하시오
    - 몬테 카를로 방법 이용하기
- cf\) 몬테 카를로 방법
    - 반복된 무작위 추출(repeated random sampling)을 이용하여 함수의 값을 수리적으로 근사하는 알고리즘
    - ![img](https://upload.wikimedia.org/wikipedia/commons/thumb/8/84/Pi_30K.gif/440px-Pi_30K.gif)

```python
import numpy as np

def automatic_lotto_number():
    result = np.random.choice(np.arange(1, 46),size = 6, replace=False)
    return result

my_lotto = automatic_lotto_number()
my_lotto
```

```python
import numpy as np

def monte_carlo_method_pi():
    total = int(1e7)
    points = np.random.rand(total, 2)
    p = 4 * np.sum(np.sum(points**2, axis = 1) <= 1) / total
    return p

monte_carlo_method_pi()
```
