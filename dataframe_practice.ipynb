## `np.random.random(size)` (alias of `np.random.random_sample()`)

- 연속형 균일분포(continuous uniform distribution)를 이용하여 0 이상 1 미만인 범위에서 샘플링된 임의의 실수를 반환합니다. size를 지정하지 않으면 값 하나를, size를 지정하면 해당하는 차원을 갖는 넘파이 배열을 생성합니다.
- `size`는 1차원의 경우 int로, 2차원 이상의 경우 **tuple**

````python
basic_rand_sample = random.random(5)
basic_rand_sample
````

````python
basic_rand1 = random.random_sample((5,5))
basic_rand1
````

## `np.random.rand(shape)`

- Create an array of the given shape and popluate it with random samples from a uniform distribution over `[0, 1)`
- `np.random.random_sample`과 유일한 차이는 shape를 tuple로 아닌 parameter 하나로 받음
- 다음은 동일한 표현이다
    - `np.random.random_sample((4,5))` (alias: `np.random.random((4,5))`)
    - `np.random.rand(4,5)`

````python
basic_rand2 = random.rand(5,5)
basic_rand2
````

import pandas as pd

}

#### 작업

- 월별 평균 기온과 강수량 계산
- 이동평균(rolling mean)을 사용하여 7일 평균 기온 트렌드 분석
- 기온과 강수량의 상관관계 분석
- 계절별(봄/여름/가을/겨울) 통계 계산

````python
import pandas as pd
import numpy as np

# 2023년 1월 1일부터 12월 31일까지의 일별 온도 데이터 생성
dates = pd.date_range(start='2023-01-01', end='2023-12-31')
temp = np.random.normal(loc=15, scale=10, size=len(dates))
rainfall = np.random.exponential(scale=5, size=len(dates))

data = {
    'date': dates,
    'temperature': temp,
    'rainfall': rainfall
}
````

````python
import matplotlib.pyplot as plt
````

````python
df_calendar = pd.DataFrame(data)
df_calendar.set_index('date', inplace=True)
````

````python
df_calendar['temp_MA7'] = df_calendar['temperature'].rolling(window = 7).mean()
````

````python
df_calendar['temp_MA7'] 
````

````python
temp_analysis = df_calendar[['temperature', 'temp_MA7']]
print("\n=== 7일 이동평균 트렌드 ===")
print(temp_analysis.head(10))
````

````python
temp_volatility = df_calendar['temperature'].rolling(window=7).std()
temp_volatility
````

````python
# 전체 기간 상관계수
correlation = df_calendar['temperature'].corr(df_calendar['rainfall'])
correlation
````

````python
# 월별 상관관계 분석
monthly_corr = df_calendar.groupby(df_calendar.index.month).apply(
    lambda x: x['temperature'].corr(x['rainfall'])
)

print("\n=== 상관관계 분석 ===")
print(f"전체 기간 상관계수: {correlation:.3f}")
print("\n월별 상관계수:")
print(monthly_corr)
````

````python
def get_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    else:
        return '겨울'

# 계절 컬럼 추가
df_calendar['season'] = df_calendar.index.month.map(get_season)

# 계절별 통계 계산
seasonal_stats = df_calendar.groupby('season').agg({
    'temperature': ['mean', 'std', 'min', 'max'],
    'rainfall': ['mean', 'sum', 'min', 'max']
})

print("\n=== 계절별 통계 ===")
print(seasonal_stats)
````

````python

# 계절별 일평균 강수량
daily_rainfall_by_season = df_calendar.groupby('season')['rainfall'].mean()
````

## Warning Message
- DataFrame에서 `inplace=True` 옵션 사용 시 나오는 경고문구
    - 추후 변경될 것이라고 하니 미리 적응해야겠다
> A value is trying to be set on a copy of a DataFrame or Series through chained assignment using an inplace method.   
> The behavior will change in pandas 3.0. This inplace method will never work because the intermediate object on which we are setting values always behaves as a copy.   
> For example, 
>   when doing 

```python 
    df[col].method(value, inplace=True)
```

> try using 

```python
df.method({col: value}, inplace=True)
#or 
df[col] = df[col].method(value)
```

> instead, to perform the operation inplace on the original object.

````python
def drop_outliers_iqr(df: pd.DataFrame, column: str) -> pd.DataFrame:
    if column not in df.columns:
        print("Column doesn't exist")
        return df
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = (q3 - q1) * 1.5
    df_drop_outliers = df[(df[column] <= (q3 + iqr)) & (df[column] >= (q1 - iqr))]

    return df_drop_outliers
````

````python
def perform_rfm_analysis(df, date_column='last_purchase'):
    # 현재 날짜 설정
    current_date = df[date_column].max()
    
    # RFM 지표 계산
    rfm = pd.DataFrame()
    rfm['recency'] = (current_date - df[date_column]).dt.days
    rfm['frequency'] = df['frequency']
    rfm['monetary'] = df['purchase_amount']
    
    # 각 지표를 5점 척도로 변환
    r_labels = range(5, 0, -1)
    f_labels = range(1, 6)
    m_labels = range(1, 6)
    
    r_quartiles = pd.qcut(rfm['recency'], q=5, labels=r_labels)
    f_quartiles = pd.qcut(rfm['frequency'], q=5, labels=f_labels)
    m_quartiles = pd.qcut(rfm['monetary'], q=5, labels=m_labels)
    
    rfm['R'] = r_quartiles
    rfm['F'] = f_quartiles
    rfm['M'] = m_quartiles
    
    # RFM 점수 계산
    rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
    
    # 고객 세그먼트 정의
    def segment_customers(row):
        if row['R'] >= 4 and row['F'] >= 4 and row['M'] >= 4:
            return 'VIP'
        elif row['R'] >= 3 and row['F'] >= 3 and row['M'] >= 3:
            return 'Loyal'
        elif row['R'] >= 3 and row['F'] >= 1 and row['M'] >= 2:
            return 'Active'
        elif row['R'] >= 2:
            return 'At Risk'
        else:
            return 'Lost'
    
    rfm['Customer_Segment'] = rfm.apply(segment_customers, axis=1)
    
    return rfm

rfm_results = perform_rfm_analysis(df)
````
